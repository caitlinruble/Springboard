## Part 2: Building and Evaluating a Model:

Build a predictive model to help Ultimate determine whether or not a user will be active in their 6th month on the system. Discuss why you chose your approach, what alternatives you considered, and any concerns you have. How valid is your model? Include any key indicators of model performance.

At it's core, this is a binary classification problem: based on the measurable factors we know about each driver, we need to build a model that will predict whether they'll still be active during their 6th month in the system.

I've already created the target variable "active" that we'll want to predict based on the other features in df. 

To prepare the data for modeling, first I'll need to dummy-encode the categorical values and scale the resulting dataframe by the standard scaler. Then, I'll split the data into a training and validation set on an 80:20 split.

I tend to like to try a few models to select which performs best in cross-validation. 
For this problem it makes sense to try a Logistic Regression model (of course!), and I'll also see how a Random Forest and XGBoost model do with the data. I'll be tuning hyperparameters and selecting the best model based on the roc-auc score using Optuna, a Bayesian hyperparameter tuning API. Optuna is a powerful platform capable of searching the hyperparameter space and selecting the optimal (or, near-optimal) model automatically, once the study is set up properly. 

I will validate the best model selected by Optuna by measuring the performance with a held-out set of test data composed of 20% of the original dataframe. I'll report the validated model performance and include a handful of useful KPIs.

```python
#one-hot encode whole df to prepare for machine learning modeling
df = pd.get_dummies(df)

#drop datetime cols to prepare for machine learning
df = df.drop(columns=['last_trip_date','signup_date'])

df.head()
```

```python
#import Standard Scaler to scale numerical features
from sklearn.preprocessing import StandardScaler

#import train_test_split
from sklearn.model_selection import train_test_split

#begin by splitting df into X and y:
X = df.drop(columns='active')
y = df['active']

#scale features using StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


# split X, y into training and testing sets, shuffling and stratifying the data based on the target variable
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, shuffle=True, stratify=df[['active']])
```

```python
print(X_scaled)
```

```python
#import modeling libraries
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

#import model selection libraries
from sklearn.model_selection import cross_val_score

#import scoring metrics and visualizers
from sklearn.metrics import roc_auc_score, plot_roc_curve, plot_precision_recall_curve,  confusion_matrix,  classification_report, accuracy_score, precision_score, recall_score, f1_score, plot_confusion_matrix
from sklearn.metrics import RocCurveDisplay, PrecisionRecallDisplay, ConfusionMatrixDisplay
from sklearn.metrics import make_scorer

#import optuna for bayesian hyper-parameter optimization
import optuna
```

```python
#dampen verbosity on optuna trial output (trust me)
optuna.logging.set_verbosity(optuna.logging.WARNING)


#set up optuna trial to select the best model
def objective(trial):
    #define which models we'll try
    classifier_name = trial.suggest_categorical("classifier", ["LogReg", "RandomForest", "XGBoost"])
    #define parameter space for Logistic Regression model
    if classifier_name == "LogReg":
        logreg_penalty = trial.suggest_categorical("logreg_penalty",  ["none", "l2"])
        logreg_c = trial.suggest_float("logreg_c", 1e-10, 1e10, log=True)
        classifier_obj = LogisticRegression(penalty = logreg_penalty, C=logreg_c, random_state=42)

    #define parameter space for XGBoost
    elif classifier_name == "XGBoost":
        xgb_gamma = trial.suggest_float("xgb_gamma", 0.1, 1.0)
        xgb_maxd = trial.suggest_int("xgb_maxd", 3, 12)
        xgb_lambda = trial.suggest_float("xgb_lambda", 1e-2, 1e0, log=True)
        xgb_alpha = trial.suggest_float("xgb_alpha", 0, 1.0)
        classifier_obj = XGBClassifier(gamma = xgb_gamma, max_depth=xgb_maxd, reg_alpha=xgb_alpha, reg_lambda=xgb_lambda, use_label_encoder=False, eval_metric='mlogloss')
    
    #define parameter space for Random Forest
    else:
        rf_max_depth = trial.suggest_int("rf_max_depth", 2, 32)
        rf_max_features = trial.suggest_int("rf_max_features", 1, 13)
        rf_n_estimators = trial.suggest_int("rf_n_estimators", 2, 32)
        classifier_obj = RandomForestClassifier(
            max_depth=rf_max_depth, n_estimators=rf_n_estimators, max_features=rf_max_features, random_state=42)

    score = cross_val_score(classifier_obj, X_train, y_train, n_jobs=-1, cv=5, scoring='roc_auc')
    auroc = score.mean()
    return auroc

if __name__ == "__main__":
    study = optuna.create_study(direction="maximize")
    study.optimize(objective, n_trials=200, n_jobs=-1)
    print(study.best_trial)
```

```python
#define best_model using the best model from the optuna study, above--an XGBoostClassifier model with average cross validation auc = 0.858
best_model = XGBClassifier(gamma = 0.70173020822384, max_depth=4, reg_alpha=0.7526788859242239, reg_lambda=0.6724087435665814, use_label_encoder=False, eval_metric='mlogloss')

#train best_model on entire training set:
best_model.fit(X_train,y_train)
#
#get predictions
y_pred = best_model.predict_proba(X_test)
y_pred_class = best_model.predict(X_test)

```

```python
print('y_pred: {}'.format(y_pred))
print('y_pred_class: {}'.format(y_pred_class))
print('y_test: {}'.format(y_test.to_numpy()))

#calculate and print KPIs
print('Final Model KPIs:')
print("AUC score: {}".format(roc_auc_score(y_test, y_pred[:,1])))
print("Accuracy score: {}".format(accuracy_score(y_test, y_pred_class)))
print("F1 score: {}".format(f1_score(y_test, y_pred_class)))
print("Precision: {}".format(precision_score(y_test, y_pred_class)))
print("Recall: {}".format(recall_score(y_test, y_pred_class)))

```

Now I need to interpret the model results in order to give actionable insights to Ultimate. Had I chosen a less accurate LogReg model, this step would have been easier, as in linear models the impacts of each feature are easier to interpret. Because I chose a tree-based model, interpretation becomes a bit more challenging.

I'll begin by plotting the feature importances:

```python
from operator import truediv
from xgboost import plot_importance, plot_tree

#reassign feature names to xgboost model
features = list(X.columns)
best_model.get_booster().feature_names = features

fig, (ax1, ax2, ax3) = plt.subplots(3,1, figsize=(7,20))

#"weight" is the number of times a feature appears in a tree
plot_importance(best_model, ax=ax1, title='Feature importance - Weight', importance_type='weight')
#"gain" is the average gain of splits which use the feature
plot_importance(best_model, ax=ax2, title='Feature importance - Gain', importance_type='gain')
#"cover" is the average coverage of splits which use the feature where coverage is defined as the number of samples affected by the split
plot_importance(best_model, ax=ax3, title='Feature importance - Cover', importance_type='cover')
plt.show
```

The different ways of calculating the feature importances give inconsistent results on which features are most important. Additionally, there is no way to tell from the feature importances *how* a feature impacts the active status. To really interpret these, we need to take a look at the SHAP values.

```python
import shap
```

```python
shap_values = shap.TreeExplainer(best_model).shap_values(X_scaled)
shap.summary_plot(shap_values, X_scaled, feature_names=features)
```

## Part 3: Recommendations to Ultimate

The SHAP value summary plot already gives us a lot of insight. Some clear insights we can glean are:

1. ***Weekday v. Weekend Use matters***:
    - Using Ultimate exclusively on weekends *or* exclusively on weekdays was negatively associated with active status at 6 months. Having an intermediate weekday percentage value was positively associated with active users at 6 months. Note: a slight exception is that some users who exclusively used Ultimate on weekdays were absolutely still active at 6 months; perhaps these are users who use Ultimate for essential weekday functions and need to continue to do so?
    - **Recommendation**: Ultimate should focus on finding incentives to encourage users to ride on weekdays *and* weekends--perhaps this looks like offering weekday ride incentives to mostly-weekend users and weekend incentives (or even event advertising) to encourage weekday riders to use Ultimate on weekends.


2. ***City Matters***:
    - Being in King's Landing is highly predictive of a user still being active at 6 months; being in Astapor made a user less likely to be active, and being in Winterfell didn't seem to matter too much in our modeling.
    - **Recommendation**: Increase advertisement to get more user signups in King's Landing to captitalize on this association. Investigate why users may be having a sub-par experience in Astapor. Do we need more drivers there? Is traffic terrible there? etc. Investigate, then address.


3. ***Phone matters***: 
    - iPhone users were no more or less likely to be active, however being an Android user is distinctly predictive of a user not being active after 6 months.
    - **Recommendation**: Ultimate should investigate any differences between the iPhone and Android app. Take steps to ensure the same app experience regardless of platform.

4. ***Initial Use Matters***
    - Users who took the highest number of trips during their first 30 days tended to be less likely to be active users in 6 months
    - **Recommendation**: evaluate how we can encourage users to "slow their roll," and enjoy the benefits of Ultimate over time rather than going down fast and hard. Are we offering discounts/incentives during that first month? Can we stretch these out over time to encourage users to adopt using Ultimate as a long-term habit rather than a short-term novelty?

5. ***Being an Ultimate Black user matters***
    - not surprisingly, Ultimate Black users were more likely to be active in the model. I would question whether they are active because they're Ultimate Black users, or whether they're Ultimate Black users because they're active. 
    - ***Recommendation***: encourage signing up for Ultimate Black as a way to encourage long-term adoption of Ultimate usage.
